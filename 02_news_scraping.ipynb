```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin
from IPython.display import Image, display

#2-1

section_dict = {
    "n0100" : '최신뉴스',
    "n0200" : '정치',
    "n0300" : '경제',
    "n0400" : '사회',
    "n0500" : '세계',
    "n0600" : 'IT/과학',
}


def print_news(sid):  # print_new(103,'생활/문화')
    #section = section_dict[sid]
    section = section_dict.get(sid)
    if section is None:
        print('요청하신 Key값이 없습니다.')
    else :
        if sid == 'n0100':
            url = f"https://news.nate.com/recent?mid=n0100"
        else:
            url = f"https://news.nate.com/section?mid={sid}"  # .format(**req_param)
        print(f"=====> {url} {section} 뉴스 <=====")

        # 요청 헤더 설정 : 브라우저 정보
        req_header = {
            "user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/104.0.0.0 Safari/537.36"
        }

        # requests 의 get() 함수 호출하기
        res = requests.get(url, headers=req_header)
        
        if res.ok:
            # res.encoding = 'utf-8'
            # 응답 (response)에서 text 추출 - source 보기
            html = res.text
            # BeautifulSoup 객체 생성  
            soup = BeautifulSoup(html, 'html.parser')
            # CSS 선택자
            a_tag_list = soup.select('div.mlt01')
            for li_tag in a_tag_list:
                a_tag = li_tag.find('a')
                link = a_tag['href']	
                    
                strong_tag = li_tag.select_one('h2.tit')
                title = strong_tag.text
                
                img_tag = li_tag.select_one('a span.ib img')
                img = img_tag['src']
                
                if img is None:
                    print(title)
                    print(link)
                else:
                    print(img)
                    print(title)
                    print(link)

                
        else:
            # 응답(response)이 Error 이면 status code 출력    
            print(f'Error Code = {res.status_code}')
    
# 함수호출

print_news('n0200')
```

```python
import requests
from bs4 import BeautifulSoup
import os

# 2-2

def download_one_episode(title, no, url):
    req_header = {
        'referer': url
    }
    res = requests.get(url)
    if res.ok:
        # jpg 파일의 절대경로 url을 찾기
        soup = BeautifulSoup(res.text, 'html.parser')
        
        img_tags = soup.select("img[src*='IMAG01']")
        img_url_list = list() # []
        for img_tag in img_tags:
            # print(type(img_tag), img_tag.name, img_tag)
            img_src = img_tag['src']
            img_url_list.append(img_src)
        #img 디렉토리 생성하기
        #img 디렉토리가 없으면 생성하기
        imgdir_name = f'img\{title}\{no}'
        if not os.path.isdir(imgdir_name):
            os.mkdir(imgdir_name)
            
        #os.path.join(dir, file)
        for img_url in img_url_list:
            # requests 의 get(url, headers) 함수 호출하기 
            res1 = requests.get(img_url, headers=req_header)
            if res1.ok:
                # binary 응답 데이터 가져오기
                img_data = res1.content
                # url에서 파일명만 추출하기
                file_path = os.path.join(imgdir_name, os.path.basename(img_url))
                # file_name = os.path.basename(img_url)
                # print(file_name)
                # binday data를 file에 write하기
                with open(file_path, 'wb') as file:
                    print(f'Writing to{file_path}({len(img_data):,} bytes)')
                    file.write(img_data)
            else:
                print(f'Error Code = {res.status_code}')
        

    else:
        print(f'Error Code = {res.status_code}')
    
download_one_episode('낢이사는이야기', 48, 'https://comic.naver.com/webtoon/detail?titleId=833255&no=49&week=tue')
```
